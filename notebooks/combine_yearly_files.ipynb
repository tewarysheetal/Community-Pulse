```python
import pandas as pd
import os
import glob
import re
```

```python
source_dir = r"D:\Projects\Campus-Community-Compact-Project\data\yearly-data"
output_dir = r"D:\Projects\Campus-Community-Compact-Project\data\combined-data"

os.makedirs(output_dir, exist_ok=True)
```

```python
for folder_name in os.listdir(source_dir):
    folder_path = os.path.join(source_dir, folder_name)

    if not os.path.isdir(folder_path):
        continue

    csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
    df_list = []

    for file in csv_files:
        filename = os.path.basename(file)

        match = re.search(r"(2019|2020|2021|2022|2023)", filename)
        if match:
            year = int(match.group(1))

            df = pd.read_csv(file)
            df.columns = df.columns.str.strip()
            df["Year"] = year
            df_list.append(df)

    if df_list:
        combined_df = pd.concat(df_list, ignore_index=True)

        output_file = os.path.join(output_dir, f"{folder_name}_combined.csv")
        combined_df.to_csv(output_file, index=False)

        print(f"Created: {output_file}")
    else:
        print(f"No matching CSV files found in {folder_name}")

print("Done")
```

```python
from pathlib import Path

output_dir = Path(r"D:\Projects\Campus-Community-Compact-Project\data\combined-data")

for file_path in output_dir.glob("*_combined.csv"):
    with open(file_path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    if len(lines) < 2:
        print(f"Skipped {file_path.name}, not enough lines")
        continue

    new_lines = lines[1:]
    new_lines[0] = new_lines[0].replace("!!", "_")
    new_lines[0] = new_lines[0].replace(" ", "_")

    with open(file_path, "w", encoding="utf-8", newline="") as f:
        f.writelines(new_lines)

    print(f"Updated: {file_path.name}")

print("Done")
```
